```python
import spacy 
nlp = spacy.load('en_core_web_sm')#'en_core_web_md'

nlp
```

```python
def random_tweet(df):
    i = np.random.choice(df['id'])
    txt = df.set_index('id').loc[i,'text']
    print(f'Tweet {i}:\n\t"{txt}"\n')
    return {i:txt} 

```

```python
## Grab 20 random tweets $$[x for y in collection for x in y]
i_tweets = []
for i in range(20):
    i_tweets.append(random_tweet(df))
    
```

```python
doc = nlp('Virtually every Democrat we are dealing with today strongly supported a Border Wall or Fence. It was only when I made it an important part of my campaign, because people and drugs were pouring into our Country unchecked, that they turned against it. Desperately needed!')
doc
```

```python
def show_entities(doc):
    if isinstance(doc,str):
        doc = nlp(doc)
        
    if doc.ents:
        for ent in doc.ents:
            print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
    else:
        print('No named entities found.')
```

```python
show_entities(doc)
```

```python
def map_get_ents(txt):
    doc = nlp(txt)
    return doc.ents

df['NER'] = df['text'].map(map_get_ents)
```

```python
df['# Ents']= df['NER'].map(len)
df['# Ents'].value_counts()
```
